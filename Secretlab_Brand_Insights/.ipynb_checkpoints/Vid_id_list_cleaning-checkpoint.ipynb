```python
import pandas as pd
import numpy as np
import re
```

```python
# Video_ids collected using Youtube Search API
vid_id_list = pd.read_csv('videoidlist_main1.csv')
# vid_id_list = pd.read_csv('videoidlist_CFAQAA.csv')
```

```python
vid_id_list.info()
```

```python
# filtering only videos that contain 'Secretlab' in their title
mask1 = (vid_id_list.video_title.str.contains('Secretlab', flags=re.IGNORECASE)) | (vid_id_list.video_title.str.contains('Secret lab',flags=re.IGNORECASE))
# removing videos that refer to the board game 'The Secret Lab'
mask2 = ~(vid_id_list.video_title.str.contains('The Secret Lab', flags=re.IGNORECASE)) | (vid_id_list.video_title.str.contains('secretlab game', flags=re.IGNORECASE))
cleaned = vid_id_list[mask1 & mask2]
```
